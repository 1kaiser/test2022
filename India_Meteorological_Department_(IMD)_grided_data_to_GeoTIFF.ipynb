```python
import imdlib as imd

# Downloading 8 years of rainfall data for India
start_yr = 2010
end_yr = 2020
variable = 'tmax' # other options are ('tmin'/ 'tmax'/'rain')
file_dir = (r'/content/data/') #Path to save the files
!mkdir -p {file_dir} #making directory from 
for i in range(start_yr, end_yr+1):
  imd.get_data(variable, i, i, fn_format='yearwise', file_dir=file_dir) #<<<< downloading dataset from 
  data = imd.open_data(variable, i, i,'yearwise', file_dir) # Opeining the downloaded dataset
  data.to_geotiff('test'+str(i)+'.tif', file_dir) #<<< convesrsion of data to geotiff
```
